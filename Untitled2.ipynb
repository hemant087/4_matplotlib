```python
plt.figure(figsize=(12,6))
bars = plt.bar(df.batter, df.runs)
sm = sum(df.runs)
for bar in bars: 
    height = bar.get_height()
    plt.annotate(f"{np.round((height/sm)*100,2)}%", xy= (bar.get_x()+bar.get_width()/2,height), va='bottom',ha="center")  
plt.xticks(rotation=45)
plt.show()
```

```python
plt.bar(df.batter,df.strike_rate,bottom=df.avg, label="Strik")
plt.bar(df.batter,df.avg, label="Average")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
