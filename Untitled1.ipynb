# 1. Grouped Bar Charts
Used when comparing multiple categories side-by-side.

```python
import matplotlib.pyplot as plt
import numpy as np

labels = ['A', 'B', 'C']
men = [20, 34, 30]
women = [25, 32, 34]

x = np.arange(len(labels))  # label locations
width = 0.35  # width of the bars

plt.bar(x - width/2, men, width, label='Men')
plt.bar(x + width/2, women, width, label='Women')

plt.xlabel('Groups')
plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(x, labels)
plt.legend()
plt.tight_layout()
plt.show()
```

# 2. Stacked Bar Charts
Stack values on top of each other for total visualization.

```python
labels = ['A', 'B', 'C']
part1 = [10, 15, 20]
part2 = [20, 25, 10]

x = np.arange(len(labels))

plt.bar(x, part1, label='Part 1')
plt.bar(x, part2, bottom=part1, label='Part 2')

plt.title('Stacked Bar Chart')
plt.xticks(x, labels)
plt.legend()
plt.show()
```

# 3. Horizontal Bar Chart

```python
categories = ['Cat A', 'Cat B', 'Cat C']
values = [3, 7, 5]

plt.barh(categories, values, color='skyblue')
plt.xlabel('Value')
plt.title('Horizontal Bar Chart')
plt.show()
```

# 4. Bar Chart with Annotations (Value Labels)

```python
values = [10, 20, 15]
labels = ['X', 'Y', 'Z']

bars = plt.bar(labels, values)

# Annotate bars
for bar in bars:
    height = bar.get_height()
    plt.annotate(f'{height}',
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 3),  # offset
                 textcoords="offset points",
                 ha='center', va='bottom')

plt.title("Bar Chart with Labels")
plt.show()
```

#  5. Custom Colors and Patterns

```python
colors = ['#FF5733', '#33FF57', '#3357FF']
hatch_patterns = ['/', '\\', 'x']

bars = plt.bar(['A', 'B', 'C'], [5, 7, 6], color=colors)

# Apply hatches
for bar, hatch in zip(bars, hatch_patterns):
    bar.set_hatch(hatch)

plt.title('Bar Chart with Custom Colors and Patterns')
plt.show()
```

# 6. Bar Chart with Error Bars

```python
values = [20, 35, 30]
errors = [2, 3, 4]

plt.bar(['A', 'B', 'C'], values, yerr=errors, capsize=5)
plt.title("Bar Chart with Error Bars")
plt.show()
```

#  7. Bar Chart from Pandas DataFrame

```python
import pandas as pd

data = {
    'Subject': ['Math', 'Physics', 'Chemistry'],
    'Hemant': [90, 85, 88],
    'Aman': [80, 89, 86]
}
df = pd.DataFrame(data)

df.set_index('Subject').plot(kind='bar')
plt.title("Student Scores by Subject")
plt.ylabel("Score")
plt.xticks(rotation=0)
plt.show()
```

# Method 1: Using ax.bar_label() (Matplotlib 3.4+)

```python
labels = ['A', 'B', 'C']
values = [10, 20, 15]

fig, ax = plt.subplots()
bars = ax.bar(labels, values, color='skyblue')

# Show value on top of each bar
ax.bar_label(bars, padding=3)

ax.set_title("Bar Chart with Values on Top")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40]
y = [5, 15, 25, 35]

plt.scatter(x, y)
plt.title("Basic Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40]
y = [5, 15, 25, 35]
sizes = [50, 100, 200, 300]  # bubble sizes
colors = ['red', 'green', 'blue', 'purple']

plt.scatter(x, y, s=sizes, c=colors, alpha=0.6, edgecolors='black')
plt.title("Advanced Scatter: Color & Size")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(50)
y = np.random.rand(50)
colors = x + y  # any metric to define color

plt.scatter(x, y, c=colors, cmap='viridis', s=100)
plt.colorbar(label='x + y value')
plt.title("Colormap Scatter Plot")
plt.show()
```

```python
x = [10, 20, 30]
y = [15, 25, 10]
labels = ['A', 'B', 'C']

plt.scatter(x, y, color='teal', s=150)

# Annotate each point
for i in range(len(x)):
    plt.annotate(labels[i], (x[i], y[i]), textcoords="offset points", xytext=(5,5), ha='center')

plt.title("Scatter Plot with Annotations")
plt.grid(True)
plt.show()
```

```python
group1_x = [1, 2, 3]
group1_y = [4, 5, 6]

group2_x = [2, 3, 4]
group2_y = [1, 2, 3]

plt.scatter(group1_x, group1_y, color='blue', label='Group 1')
plt.scatter(group2_x, group2_y, color='orange', label='Group 2')

plt.legend()
plt.title("Multiple Groups Scatter Plot")
plt.show()
```
