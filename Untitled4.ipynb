```python
# Full pyramid
def full_pyramid(rows):
    for i in range(rows):
        print(" " * (rows - i - 1) + "*" * (2 * i + 1))

# Inverted pyramid
def inverted_pyramid(rows):
    for i in range(rows, 0, -1):
        print(" " * (rows - i) + "*" * (2 * i - 1))

# Hollow pyramid
def hollow_pyramid(rows):
    for i in range(rows):
        if i == 0 or i == rows - 1:
            print(" " * (rows - i - 1) + "*" * (2 * i + 1))
        else:
            print(" " * (rows - i - 1) + "*" + " " * (2 * i - 1) + "*")

# Half pyramid
def half_pyramid(rows):
    for i in range(rows):
        print("*" * (i + 1))

# Inverted half pyramid
def inverted_half_pyramid(rows):
    for i in range(rows, 0, -1):
        print("*" * i)

# Number pyramid
def number_pyramid(rows):
    for i in range(1, rows + 1):
        print(" " * (rows - i), end="")
        for j in range(1, i + 1):
            print(j, end=" ")
        print()

# Example usage
full_pyramid(5)
inverted_pyramid(5)
hollow_pyramid(5)
half_pyramid(5)
inverted_half_pyramid(5)
number_pyramid(5)
```
